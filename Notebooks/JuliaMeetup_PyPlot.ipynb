## Simple Examples

Here we avoid all extraneous features and just show some of the basic plot types that can be created with PyPlot.jl

```julia
npts = 125

d = randn(npts)
x = linspace(0, 2*Ï€, npts)
y = copy(x)
sinx = sin(x)
cosx = cos(x)
z = cos(x') .+ sin(y);
```

### Line Plot

```julia
plot(x, sinx)
```

### Scatter Plot

```julia
scatter(x, d)
```

### Histogram

```julia
plt[:hist](d);
```

Why do we need the `plt[:hist]`? Reading the docs will warn you that:

> You must also use `plt` to access some functions that conflict with built-in Julia functions. In particular `plt[:hist]`, `plt[:xcorr]`, and `plt[:isinteractive]` must be used to access `matplotlib.pyplot.hist etcetera`.

### Heatmap

```julia
pcolormesh(x, y, z)
```

### Surface Maps

`PyPlot.jl` makes our life easier than in `matplotlib` because we aren't required to create a `Axes3D` object before making a 3-D plot

```julia
surf(x', y, z)
```

## More Features

Build up 4 plots using different features

```julia
fig, ax = subplots(2, 2)

ax1, ax2, ax3, ax4 = ax
tight_layout()
```

### Visualization Features

```julia
ax1[:plot](x, sinx, color="k", linewidth=5.0, linestyle="-", alpha=0.3)
ax1[:scatter](x[1:5:end], sinx[1:5:end], color=["red", "green", "blue"],
              s=100*abs(sinx[1:5:end]))
ax3[:hist](d, color="#009900")
fig
```

```julia
get_cmaps()[1:5:end]
```

```julia
ax2[:pcolormesh](z, cmap="bwr")
ax4[:barh](x[1:5:end], cosx[1:5:end], color=["red", "blue", "green"])
fig
```

### Titles and Axis Labels

```julia
ax1[:set_title]("Sin plot with colored dots")
ax3[:set_title]("Green Histogram")
ax1[:set_ylabel]("sin(x)")
ax3[:set_ylabel]("Density")
fig
```

### Change Axis Limits

```julia
ax1[:set_xlim](0, 2*Ï€)
ax2[:set_xlim](1, npts)
ax2[:set_ylim](1, npts)
fig
```

### Annotation

```julia
littleindex = div(npts, 2)
xlittle, sinlittle = x[littleindex-1], sinx[littleindex-1]
ax1[:annotate](L"$\min_x sin(x)$", xy=(xlittle, sinlittle),
               xytext=(xlittle+1, sinlittle+0.5),
               arrowprops=Dict(:facecolor=>"black", :shrink=>0.1))
fig
```

### Facecolors and backgrounds

```julia
ax1[:set_axis_bgcolor]("#ccccff")
ax3[:set_axis_bgcolor]("magenta")
ax4[:set_axis_bgcolor]((255/256, 255/256, 153/256))
fig[:set_facecolor]("blue")
fig
```

## Plots.jl

Unified interface to various plotting packages

Add `Plots.` in front of all commands because it conflicts with previous namespace from `PyPlot`. Wouldn't normally need that prefix.

```julia
using Plots
pyplot()
```

```julia
Plots.plot(x, sinx, leg=false, color=:red, style=:dash)
```

```julia
Plots.contour(x, y, z, fill=true)
Plots.scatter!([2.5], [2.5], color=:black, ms=3, leg=false)
Plots.annotate!([(2.5, 2.5, Plots.text("This is 2.5, 2.5", 12, :white, :left))])
```

```julia
Plots.heatmap(randn(10000), randn(10000), nbins=25)
Plots.title!("Heatmap")
```

```julia
Plots.subplot(randn(100, 4), layout=[2, 2], t=[:line :hist :scatter :bar],
              nbins=25, leg=false, color=[:red :blue :green :black])
```
